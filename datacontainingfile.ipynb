```python
import pandas as pd

df = pd.read_csv("input_table.csv")
df.head()
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')  # lightweight and fast

# Create embeddings for each row (combine relevant fields)
df['combined_text'] = df.apply(lambda row: f"{row['Product line']} {row['gross income']}", axis=1)
df['embedding'] = df['combined_text'].apply(lambda x: model.encode(x))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Example Query
query = "What is the highest gross income for health and beauty product line?"
query_embedding = model.encode(query)

# Compute similarity
df['similarity'] = df['embedding'].apply(lambda x: cosine_similarity([x], [query_embedding])[0][0])

# Get top N relevant rows
top_n = df.sort_values(by='similarity', ascending=False).head(5)
top_n[['Product line', 'gross income', 'similarity']]
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

prompt = "What is the average gross income for electronic accessories?"  # <-- Add this!

# Tokenize
inputs = tokenizer(prompt, return_tensors="pt")
print("Input Token IDs:", inputs['input_ids'])

# Generate output token IDs
outputs = model.generate(**inputs)
print("Output Token IDs:", outputs)

# Decode output tokens to string
decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Final Output:", decoded)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd

# Load question sheet
df = pd.read_excel("myquery.xlsx")

# Load FLAN-T5 model
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

generated_responses = []

# For each question, generate response
for q in df["question"]:
    inputs = tokenizer(q, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=100)
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated_responses.append(decoded)

# Add to dataframe
df["generated response"] = generated_responses

# ✅ Overwrite original file with generated responses
df.to_excel("myquery.xlsx", index=False)

print("✅ Done! Check 'myquery.xlsx' for the results.")

```

```python
import pandas as pd

# Load the input table and queries
data_df = pd.read_csv("input_table.csv")
query_df = pd.read_excel("myquery.xlsx")

# Helper function to evaluate each question
def evaluate_query(q):
    q_lower = q.lower()

    try:
        if "latest entry" in q_lower and "product line" in q_lower:
            return data_df.iloc[-1]["Product line"]

        elif "first transaction" in q_lower:
            return data_df["Date"].min()

        elif "latest transaction" in q_lower:
            return data_df["Date"].max()

        elif "max rating" in q_lower and "home and lifestyle" in q_lower:
            filtered = data_df[data_df["Product line"].str.lower() == "home and lifestyle"]
            return filtered["Rating"].max()

        elif "male customers" in q_lower and "fashion accessories" in q_lower:
            filtered = data_df[(data_df["Gender"] == "Male") & (data_df["Product line"] == "Fashion accessories")]
            return len(filtered)

        elif "average gross income" in q_lower and "electronic accessories" in q_lower:
            filtered = data_df[data_df["Product line"] == "Electronic accessories"]
            return round(filtered["gross income"].mean(), 2)

        # Add more conditions as needed...

        else:
            return "Not handled"

    except Exception as e:
        return f"Error: {e}"

# Apply function to each question
query_df["evaluated answer"] = query_df["question"].apply(evaluate_query)

# Save the updated file
query_df.to_excel("myquery.xlsx", index=False)
print("✅ Done! 'evaluated answer' column added to myquery.xlsx")
```

```python
import pandas as pd

# Load input table and queries
input_df = pd.read_csv("input_table.csv")
query_df = pd.read_excel("myquery.xlsx")

evaluated_answers = []

# Go through each query
for idx, row in query_df.iterrows():
    try:
        # Parse row and column indexes
        row_idxs = [int(i.strip()) for i in str(row['row index']).split(',') if i.strip().isdigit()]
        col_idxs = [int(i.strip()) for i in str(row['column index']).split(',') if i.strip().isdigit()]

        # Handle case 1: single cell lookup
        if len(row_idxs) == 1 and len(col_idxs) == 1:
            evaluated = input_df.iloc[row_idxs[0], col_idxs[0]]

        # Handle case 2: multiple rows and one column (e.g., for max, min, count)
        elif len(row_idxs) > 0 and len(col_idxs) == 1:
            col_data = input_df.iloc[row_idxs, col_idxs[0]]
            # Try to infer logic from question
            q = row['question'].lower()
            if "max" in q:
                evaluated = col_data.max()
            elif "min" in q:
                evaluated = col_data.min()
            elif "average" in q or "mean" in q:
                evaluated = round(col_data.mean(), 2)
            elif "count" in q or "how many" in q:
                evaluated = len(col_data)
            else:
                evaluated = ", ".join(col_data.astype(str).unique())

        # Handle unsupported or complex logic
        else:
            evaluated = "Not handled"

    except Exception as e:
        evaluated = "Error"

    evaluated_answers.append(evaluated)

# Add to the sheet
query_df["evaluated answer"] = evaluated_answers
query_df.to_excel("myquery.xlsx", index=False)
print("✅ Evaluated answers written to 'myquery.xlsx'")
```

```python
def match(generated, evaluated):
    return "Yes" if str(generated).strip().lower() == str(evaluated).strip().lower() else "No"

df["is_match"] = df.apply(lambda row: match(row["generated response"], row["evaluated answer"]), axis=1)
df.to_excel("myquery.xlsx", index=False)
print("✅ Match column added.")
```
